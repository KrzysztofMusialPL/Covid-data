```python
import matplotlib.pyplot as plt

my_data = [39328.0, 19605.0, 1022.0, 2451.0, 1373.0, 22363.0, 2600.0, 2752.0, 12761.0, 3463.0, 4251.0]
my_labels = 'PNEUMONIA', 'DIABETES', 'ASTHMA', 'COPD', 'INMUSPR', 'HIPERTENSION', 'OTHER_DISEASE', 'CARDIOVASCULAR', 'OBESITY', 'RENAL_CHRONIC', 'TOBACCO'

fig, ax = plt.subplots(figsize =(16, 9))
 
ax.barh(my_labels, my_data)
 
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 

for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
ax.set_title('Udział chorób współistniejących w liczbie zgonów',
             loc ='center', fontsize = 20)
 
fig.text(0.9, 0.15, 'Pogromcy Danych', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)

plt.show()
```
